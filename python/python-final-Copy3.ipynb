## Q5. For which courses the percentage of male/female students is over 70%?

```python
new_df3 = pd.merge(classrooms,students, left_on=['StudentId'], right_on = ['StudentId'])
new_df3['Female'] = 0
new_df3['Male'] = 0
new_df3.loc[new_df3.Gender == 'F','Female'] = 1
new_df3.loc[new_df3.Gender == 'M','Male'] = 1
calc = new_df3.groupby('CourseId').agg({'Female':['mean']})
if (calc[calc > 0.7].dropna().shape[0] > 0 ):
    print( calc[calc > 0.7].dropna())
calc = new_df3.groupby('CourseId').agg({'Male':['mean']})
if (calc[calc > 0.7].dropna().shape[0] > 0 ):
    print( calc[calc > 0.7].dropna())
```
