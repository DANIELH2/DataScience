```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
df = pd.read_csv (r'C:\Users\DANIEL\Documents\DataScience\data\all_stocks_5yr.csv', index_col='date', parse_dates=True)
df.head()
```

```python
df_aapl=df[df.Name == 'AAPL'] # create a new dataframe with just the apple data

win_len = 30 #how long thw eindow wil be
win_type = [None, 'boxcar', 'triang', 'blackman', 'hamming', 'nuttall'] #window types 
 
ax = df_aapl['open'].plot(figsize=(7, 4), title='AAPL Open Price') ## plot the raw Open data

# loop over the window types
for win in win_type:
    df_aapl_roll = df_aapl.rolling(win_len, win_type=win).mean() #create a loop betwin win_type
    
    # rename the columns so we can see them in the legend    
    df_aapl_roll.columns = [
        '{}_{}_{}'.format(col, win_len, win) for col in df_aapl_roll.columns
    ]
    df_aapl_roll['open_{}_{}'.format(win_len, win)].plot()
    
# show everything
ax.legend(fontsize=8.5)
plt.tight_layout()#make the plot more strech
plt.show


```
