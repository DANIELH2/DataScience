```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=["Vitamin C"])
# Interest by Region
df = pytrend.interest_by_region()
df.head(10)
```

```python
df.reset_index().plot(x="geoName", y="Vitamin C", figsize=(50, 10), kind ="bar")
```

```python
#Which Programming language is more popular in the last 3 months in Taiwan?
pytrend = TrendReq(hl='en-US', tz=360)
keywords = ['Python', 'R']
pytrend.build_payload(
     kw_list=keywords,
     cat=0,
     timeframe='today 3-m',
     geo='TW',
     gprop='')
data = pytrend.interest_over_time()
data.to_csv('Py_VS_R.csv', encoding='utf_8_sig')
```

```python
data = data.drop(labels=['isPartial'],axis='columns')
data.to_csv('Py_VS_R.csv', encoding='utf_8_sig')
```

```python
image = data.plot(title = 'Python V.S. R in last 3 months on Google Trends')
fig = image.get_figure()
fig.savefig('figure.png')
```
