```r
df = read.csv("../data/Daily_Demand_Forecasting_Orders.csv" , sep = ";")
```

```r
orders <- df[,c(1,2,13)]

names(orders)<- c("week","weekday","total")
head(orders)
```

```r
orders.ts<- ts(orders$total,start = 1,end =60,frequency = 5)
head(orders.ts)
```

```r
plot(orders.ts)
```

```r
lag3<- stats::lag(orders.ts,3)
```

```r
ma5<- TTR::SMA(orders.ts,n=5)
ma3<- TTR::SMA(orders.ts,n=3)
```

```r
plot(orders.ts)
lines(lag3,col="green")
lines(ma5,col="blue")
lines(ma3,col="red")
```

```r
options(pepr.plot.width=15,repr.plot.hight=3)
plot(orders.ts)
lines(lag3,col="green")
lines(ma5,col="blue")
lines(ma3,col="red")
```
